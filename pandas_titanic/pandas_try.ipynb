```python
titanic['Fare']
```

```python
q = titanic['Fare'].quantile(0.9) #calculating 90th percentile
```

```python
titanic['Fare'] = titanic['Fare'].mask(titanic['Fare'] > titanic['Fare'].quantile(.9), titanic['Fare'].quantile(.9))
```

```python
titanic['Fare']>q
```

```python
titanic.loc[titanic['Fare']>q, 'Fare'] = q
```

```python
titanic.loc[titanic['Fare']>=q, 'Fare']
```

```python
titanic['Fare'].nlargest(100) # checking if replacing values worked xD
```
